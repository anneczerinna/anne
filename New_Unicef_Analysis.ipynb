# Wasting Rates in Children Under 5 years old

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV again if needed
data = pd.read_csv('/content/unicef_indicator_1.csv')  # adjust path if necessary

# Filter for the 3 countries and the specific indicator
countries = ['Bangladesh', 'India', 'Pakistan']
wasting_data = data[
    (data['indicator'] == 'Weight-for-height <-2 SD (wasting)') &
    (data['country'].isin(countries))
].copy()

# Convert year column to numeric
wasting_data['time_period'] = pd.to_numeric(wasting_data['time_period'], errors='coerce')

# Filter for total values only (ignoring sex split for now)
wasting_total = wasting_data[wasting_data['sex'] == 'Total']

# Now plot
plt.figure(figsize=(10,6))

for country in countries:
    subset = wasting_total[wasting_total['country'] == country]
    plt.plot(subset['time_period'], subset['obs_value'], marker='o', label=country)

plt.title('Wasting Rates Over Time (Under 5 Children)')
plt.xlabel('Year')
plt.ylabel('Wasting Rate (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Wasting Rates in Children Under 5**

Wasting is characterised by low weight-for-heightis a critical indicator of acute malnutrition, particularly among children under five. In South Asia, encompassing Bangladesh, India, and Pakistan, wasting rates have remained persistently high, posing significant challenges to child health and survival.

This graph shows wasting rates, an indicator of acute malnutrition in children under 5 across Bangladesh, India, and Pakistan over time.
- Bangladesh has seen a steady decline, dropping from around 17% to below 10% by 2020.
- India has had consistently high rates, staying around 1820%, with only slight improvement.
- Pakistan started high at over 23%, but made major progress, falling to around 7% by 2020.

This reflects varied success in tackling child malnutrition, with Bangladesh and Pakistan showing strong improvement, while India still faces challenges.

# Wasting Prevalence Overtime  

```python
import pandas as pd
import plotly.express as px

# Load data
data = pd.read_csv('/content/unicef_indicator_1.csv')

# Filter for wasting indicator
scatter_data = data[data['indicator'] == 'Weight-for-height <-2 SD (wasting)'].copy()

# Convert time_period and obs_value to numeric
scatter_data['time_period'] = pd.to_numeric(scatter_data['time_period'], errors='coerce')
scatter_data['obs_value'] = pd.to_numeric(scatter_data['obs_value'], errors='coerce')

# Drop rows with missing values in those columns
scatter_data = scatter_data.dropna(subset=['time_period', 'obs_value', 'observation_status'])

# Plot
fig = px.scatter(
    scatter_data,
    x='time_period',
    y='obs_value',
    color='obs_value',
    trendline='ols',
    trendline_color_override='black',
    # color='observation_status',  # Optional: if you want color differentiation
    hover_data=['country', 'current_age'],  # Display extra data on hover
    title='Wasting Prevalence over Time',
    labels={'time_period': 'Year', 'obs_value': 'Wasting Prevalence (%)'}
)

fig.show()
```

**Wasting Prevalence Overtime**  

Each point represents a recorded observation of wasting prevalence for a country at a specific year.

- The distribution of points over time shows variability across countries, highlighting that progress has been uneven.
- The black trend line illustrates the overall global tendency: a gradual decline in wasting prevalence from 2000 onwards.
- Persistent outliers indicate that while many countries have improved, some continue to face high rates of child wasting.

Overall, the visual demonstrates that although positive progress has been made, achieving significant and sustained reductions remains a global challenge.

# Wasting Prevalence by Country  

```python
!pip install geopandas

import pandas as pd
import plotly.express as px
import geopandas as gpd

# Load world map from Natural Earth (GeoJSON)
world = gpd.read_file('https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_110m_admin_0_countries.geojson')

# Load data
data = pd.read_csv('/content/unicef_indicator_1.csv')

# Filter for wasting indicator
latest_data = data[data['indicator'] == 'Weight-for-height <-2 SD (wasting)'].copy()

# Convert time_period to numeric
latest_data['time_period'] = pd.to_numeric(latest_data['time_period'], errors='coerce')

# Keep latest record for each country
latest_data = latest_data.sort_values('time_period').groupby('country').last().reset_index()

# Merge GeoDataFrame with wasting data
merged = world.merge(latest_data, how='left', left_on='NAME', right_on='country')

# Plot map
fig = px.choropleth(
    merged,
    geojson=merged.geometry,
    locations=merged.index,
    color='obs_value',  # Color based on wasting rate
    hover_data=['NAME', 'obs_value'],  # Display country name and wasting rate on hover
    title='Wasting Prevalence by Country (Latest Data)',
    color_continuous_scale='sunsetdark',  # Choose a color scale
    scope='world'
)

  # Focus on plotted countries
fig.show()
```

**Wasting Prevalence by Country**  

This choropleth map presents the most recent available data on child wasting prevalence across countries.

- Countries are color-coded based on the percentage of children affected, with darker shades indicating higher rates.
- Data integration combines geographic information with UNICEF's latest country-level wasting observations.
- Gaps or missing countries reflect where updated data were not available at the time of collection.

The global distribution highlights that while some regions have made significant progress, others, particularly in South Asia and parts of Africa  continue to experience alarmingly high rates of wasting.

# Wasting Prevalence Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/unicef_indicator_1.csv')

# Filter for wasting indicator and selected countries
heatmap_data = data[
    (data['indicator'] == 'Weight-for-height <-2 SD (wasting)') &
    (data['country'].isin(['Bangladesh', 'India', 'Pakistan']))
].copy()

# Convert time_period and obs_value to numeric
heatmap_data['time_period'] = pd.to_numeric(heatmap_data['time_period'], errors='coerce')
heatmap_data['obs_value'] = pd.to_numeric(heatmap_data['obs_value'], errors='coerce')

# Pivot the data: countries as rows, years as columns
pivot_table = heatmap_data.pivot_table(
    index='country',
    columns='time_period',
    values='obs_value'
)

# Plotting
plt.figure(figsize=(14, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlOrRd", linewidths=0.5, cbar_kws={'label': 'Wasting Prevalence (%)'})

plt.title('Wasting Prevalence Heatmap (Bangladesh, India, Pakistan)', fontsize=16, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

Wasting Prevalence Heatmap

- **Bangladesh:** Wasting prevalence declined from 17.3% (1986) to 8.4% (2018), with occasional peaks (e.g., 20.7% in 1997).
- **India:** Rates remained high, fluctuating between 15.1% (2014) and 20.8% (2015), with 18.7% in 2020.
- **Pakistan:** Sharp decrease from 24% (1986) to 7.1% (2018), with intermediate values like 17.2% (1991) and 14.9% (2011).

Overall, all three countries show a reduction in wasting prevalence over time, but Indias rates remain relatively high compared to Bangladesh and Pakistan.

# Conclusion

**Conclusion** 

UNICEF Report on Undernutrition in South Asia
South Asia remains the global epicenter of child wasting, with India, Bangladesh, and Pakistan accounting for a major share of the world's wasted children.

- Bangladesh: Wasting prevalence dropped from 17.3% (1986) to 8.4% (2018), showing steady improvement.
- India: Wasting rates remain persistently high, fluctuating between 15% and 21% in recent decades, with the latest estimate at 18.7%-the highest globally.
- Pakistan: Marked progress, with prevalence falling sharply from 24% (1986) to 7.1% (2018).
- Gender Disparity: Boys consistently have higher wasting rates than girls in all three countries.
- Regional Burden: Over half of the worlds wasted children live in South Asia, and prevalence here is double that of sub-Saharan Africa.
- Key Drivers: Poor maternal nutrition, low birth weight, and inadequate infant feeding practices are major contributors to high wasting rates in the region.
- Policy Gaps: Despite improvements, treatment coverage remains low, and policy responses are often insufficient or contested, especially in India and Bangladesh.

While Bangladesh and Pakistan have achieved significant reductions, India continues to face severe challenges, highlighting the urgent need for targeted interventions and sustained policy action.

South Asia has made progress in reducing child wasting, but the region-particularly India-still faces a critical burden of acute malnutrition. Addressing maternal health, improving infant feeding, and scaling up treatment are essential for further progress.
